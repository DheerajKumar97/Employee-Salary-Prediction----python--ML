```python
cat_var = [var for var in df.columns if df[var].dtypes =="object"]
cat_var = df[cat_var]
cat_var
```

```python
def encode_Age(Age):
    if Age == "0-17":
        return 1
    elif Age == "55+":
        return 2
    elif Age == "26-35":
        return 3
    elif Age == "46-50":
        return 4
    elif Age == "51-55":
        return 5
    elif Age == "36-45":
        return 6
    elif Age == "18-25":
        return 7
    else:
        return 0
cat_var['Age'] = cat_var['Age'].apply(encode_Age)
```

```python
def encode_Stay_In_Current_City_Years(Stay_In_Current_City_Years):
    if Stay_In_Current_City_Years == "4+":
        return 4
    elif Stay_In_Current_City_Years == "1":
        return 2
    elif Stay_In_Current_City_Years == "2":
        return 3
    elif Stay_In_Current_City_Years == "3":
        return 4
cat_var['Stay_In_Current_City_Years'] = cat_var['Stay_In_Current_City_Years'].apply(encode_Stay_In_Current_City_Years)
```

```python
cat_var
```

```python
cat_var['Gender'] = pd.get_dummies(cat_var['Gender'])
```

```python
def encode_City_Category(City_Category):
    if City_Category == "A":
        return 1
    elif City_Category == "B":
        return 2
    elif City_Category == "C":
        return 3
cat_var['City_Category'] = cat_var['City_Category'].apply(encode_City_Category)
```

```python
cat_var['Purchase'] = df['Purchase']
cat_var
```

```python
mean_encode = cat_var.groupby('Product_ID')['Purchase'].mean()
```

```python
cat_var.loc[:,'encoded_id'] = cat_var.Product_ID.map(mean_encode)
```

```python
mean = cat_var['Purchase'].mean()
agg =  cat_var.groupby('Product_ID')['Purchase'].agg(['count','mean'])
counts = agg['count']
means =agg['mean']
weight =100
```

```python
smooth = (counts * means + weight * means) / (counts + weight)
smooth
```

```python
cat_var.loc[:,'encoded_id'] = cat_var.Product_ID.map(smooth)
```

```python
cat_var =cat_var.drop(['Product_ID'],axis=1)
```

```python
cat_var
```

```python
num_var = [var for var in df.columns if df[var].dtypes !="object"]
num_var = df[num_var]
num_var
```

```python
#num_var = num_var.drop(['Purchase'],axis=1)
num_var = num_var.drop(['Product_Category_2'],axis=1)
num_var = num_var.drop(['Product_Category_3'],axis=1)
num_var
```

```python
df_new = pd.concat([num_var,cat_var],axis=1)
df_new
```

```python
df_new = df_new.apply(lambda x:(x.astype(float) - min(x))/(max(x)-min(x)), axis = 0)
df_new
```

```python
df_new.isna().sum()
```

```python
df_new['Stay_In_Current_City_Years'] = df_new['Stay_In_Current_City_Years'].fillna(df_new['Stay_In_Current_City_Years'].mode()[0])
```

```python
x = df_new.drop(['Purchase'],axis=1)
x = df_new.drop(['Occupation'],axis=1)
x = df_new.drop(['Stay_In_Current_City_Years'],axis=1)
y = df_new['Purchase']
```
